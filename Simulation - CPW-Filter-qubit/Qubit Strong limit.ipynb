# The full Hamiltonian of the system is:

Using the following Hamiltonian [1] to describe the cavity-qubit dynamics in the strong dispersive regime:

$\displaystyle H = \omega_r a^\dagger a - \frac{1}{2}\omega_q \sigma_z + \chi (a^\dagger a  + 1/2) \sigma_z$

where $\omega_r$ is the cavity resonace frequency, $\omega_q$ is the qubit energy and $\chi$ is the qubit-cavity dispersive interaction strength.

Also the coupling strength is:

$g = e\dfrac{C_g}{C_{\Sigma}}\dfrac{V^0_{rms}}{\hbar}$,

and $\chi = g^2/\Delta$, with $\Delta = \omega_q - \omega_r$



$\kappa$: cavity relaxation rate,($\kappa = \frac{\omega_r}{Q}$)

$\gamma$: qubit relatation rate,

$\Gamma$: qubit pump rate,

$n_{th}$: thermal occuppancy






[1] Schuster, D. I. et al. Resolving photon number states in a superconducting circuit. Nature 445, 515–518 (2007).

```python
# Definition of the function to simulate the Hamiltonian
def calc_spectrum(N,M, wr, wq, g, kappa, gamma, Gamma, n_th_a, tlist):
    #where:
        #N, M are the Hilbert space size for the cavity and NR
        # wr: cavity angular frequency
        #wq: qubit Josephson frequency
        #wqc: qubit Charge frequency
        #wm: NR frequency
        #g: cavity-qubit coupling
        #g1: qubit-NR coupling
        #kappa:  cavity dissipation rate
        #gamma:  atom dissipation rate
        #Gamma:  atom pump rate
        #Gamma_m: NR pump rate
        #n_th_a:  avg number of thermal bath excitation
        #tlist: time list

    chi = g**2/(wq-wr)
    print(chi)
    # cavity operators
    a = tensor(destroy(N), qeye(2))
    nc = a.dag() * a
    xc = a + a.dag()

    # mechanical resonator operators

#     b = tensor(qeye(N),destroy(M), qeye(2)) #tensor(thermal_dm(N, 3), ket2dm(basis(2,0)+basis(2,1))).unit() #
#     ncm = b.dag()*b
#     xcm = b + b.dag()


    # qubit operators

    sm = tensor(qeye(N), destroy(2))
    sz = tensor(qeye(N), sigmaz())
    sx = tensor(qeye(N), sigmax())
    nq = sm.dag() * sm
    xq = sm + sm.dag()
    I = tensor(qeye(N), qeye(2))


    # coupling 
    # dispersive hamiltonian
     
    Hr = wr * (a.dag() * a + I/2.0) 
    Hq = (wq / 2.0) * sz 
    Hi = chi * (a.dag() * a + I/2) * sz
#     Hm = wm * (b.dag() * b + I/2.0)
#     Hq = ((sqrt(wq**2+(8*wqc)**2)) / 2.0)*sz 
    #Hq =  (4*wqc)*sx -(wq/2.0)*sz 
    #Hi = ( g * (a.dag() + a) + g1 * (b.dag() + b) ) * (8*wqc*sz - (wq/2.0)*sx) 
    #Hi = (  (g1 * (b.dag() + b)) ) * (wqc*sz - (wq/2.0)*sx) 
#     Hi2 = g * (a.dag() + a) * (sx)
    #Hi = (  (g1 * (b.dag() + b)) ) * (wqc*sz + (wq/2.0)*sx) 
    H = Hr + Hq + Hi# +Hi2
    
    
#     psi0 = tensor(thermal_dm(N,sqrt(0.5)), ket2dm(basis(2,0))).unit()
#     psi0 = tensor((0.25*thermal_dm(N,sqrt(1))+ coherent_dm(N,sqrt(M))), ket2dm(basis(2,0))).unit()
    psi0 = tensor((coherent_dm(N,sqrt(M))), ket2dm(basis(2,0))).unit()



    #psi0 = tensor(coherent_dm(N,sqrt(3)),thermal_dm(M,sqrt(4)), ket2dm(basis(2,0))).unit()
    #psi0 = tensor(fock(N,1),fock(M,4), basis(2,0)).unit()

    # collapse operators

    c_op_list = []

      


    rate = kappa * (1 + n_th_a)
    if rate > 0.0:
        c_op_list.append(sqrt(rate) * a)

    rate = kappa * n_th_a
    if rate > 0.0:
        c_op_list.append(sqrt(rate) * a.dag())

    rate = gamma
    if rate > 0.0:
        c_op_list.append(sqrt(rate) * sm)

    rate = Gamma
    if rate > 0.0:
        c_op_list.append(sqrt(rate) * sm.dag())   
    
#     rate = Gamma_m
#     if rate > 0.0:
#         c_op_list.append(sqrt(rate) * b.dag()) 

    A = sx.dag()
    B = sx
    spec = []

    # calculate the power spectrum
    corr_vecq = correlation_2op_2t(H, psi0, None, tlist, c_op_list, A, B,solver="me")

    print("Correlation Done!")

    w1, S1 = spectrum_correlation_fft(tlist, corr_vecq)

    print( "Spectrum Done!")

    #spec = spectrum_ss(H, wlist, c_op_list, A, B)

    return w1, S1
```
